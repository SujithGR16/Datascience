```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
type(df1)
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SATCORES')
plt.ylabel('FREQUENCY')
plt.title('Historgram of SAT scores')
plt.show()
```

```python
sns.histplot(df1['SAT'])
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SAT Scores')
plt.ylabel('Frequency')
plt.title('Histogram of SAT Scores')
plt.show()
```

```python
sns.histplot(data=df1, x='SAT')
plt.xlabel('SAT Scores')
plt.ylabel('Frequency')
plt.title('Histogram of SAT Scores (Seaborn)')
plt.show()
```

```python
sns.histplot(df1['Accept'],color='purple')
plt.xlabel('Accept rate')
plt.ylabel('Frequency')
plt.title('Histogram Accept rate')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'bo')
plt.xlabel('SAT')
plt.ylabel('Accept')
plt.title('Scatter plot of SAT vs Accept')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Top10'],'rs')
plt.xlabel('SAT')
plt.ylabel('Top10')
plt.title('Scatter plot of SAT vs Accept')
plt.show()
```

```python
plt.plot(df1['SAT'], df1['Top10'], 'bo')
plt.xlabel('SAT Scores')
plt.ylabel('Top10 Percentage')
plt.title('Scatter plot of SAT vs Top10')
plt.show()
```
